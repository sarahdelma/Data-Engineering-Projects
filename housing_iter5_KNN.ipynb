```python
data = data1[['LotArea', 'TotalBsmtSF', 'BedroomAbvGr','GarageCars','Expensive']]
```

```python

#input features for model
X=data.drop(['Expensive'],axis=1)
#output features for model
y=data['Expensive']
```

```python

X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, test_size=0.2, random_state=42
)
```

Create the model

```python
 
# pipeline
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
pipe = make_pipeline(
        SimpleImputer(strategy='median'),
        StandardScaler(),
        KNeighborsRegressor())
 
# parameter grid for pipeline
pipe_params = {
    'simpleimputer__strategy':['median', 'mean'],
    'standardscaler__with_mean':[True, False],
    'kneighborsregressor__n_neighbors': list(range(1, 20)),
    'kneighborsregressor__weights': ['uniform', 'distance'],
    'kneighborsregressor__p': [1, 2],
    'kneighborsregressor__algorithm': ['ball_tree', 'kd_tree', 'brute']}
 
```

```python

# grid search
from sklearn.model_selection import GridSearchCV
trained_pipe = GridSearchCV(pipe,
                            pipe_params, 
                            cv = 5)
trained_pipe.fit(X_train, y_train)
 
# test accuracy on the test set
from sklearn.metrics import r2_score


#Predict
y_pred = trained_pipe.predict(X_test)


#Score
r2 = r2_score(y_test, y_pred)
print(r2)
```

```python
y_predicted = trained_pipe.predict(X_test)
y_predicted
```

```python
# store the trained pipeline
import pickle
pickle.dump(trained_pipe, 
            open(file='trained_pipe_knn.sav', 
                 mode='wb'))
```
